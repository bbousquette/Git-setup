```python
print("Michigan's Average Grade 4 Math Score is:")
print(df["AVG_MATH_4_SCORE"][df["STATE"] == "MICHIGAN"].mean())
print("Ohio's Average Grade 4 Math Score is:")
print(df["AVG_MATH_4_SCORE"][df["STATE"] == "OHIO"].mean())
print("Thus, Ohio's Average Grade 4 Math Score is Higher than Michigan's.")
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['ENROLL_4','AVG_MATH_8_SCORE','AVG_READING_4_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_4_SCORE']
y.fillna(y.median(), inplace=True)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
# model = 
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print("R^2: ")
print(model.score(X_test, y_test))
print("Average error: ")
print(np.mean(model.predict(X_test)-y_test))
print("Root mean squared: ")
print(np.mean((model.predict(X_test)-y_test)**2)**0.5)
```
